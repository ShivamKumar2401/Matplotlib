```python
import matplotlib.pyplot as plt
```

```python
years=[1990,1992,1994,1996,1998,2000,2003,2005,2007,2010]
runs=[500,700,1100,1500,1800,1200,1700,1300,900,1500]
```

```python
plt.bar(years,runs)
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar yearly runs")
```

```python
import numpy as np
```

```python
sachin=[100,1100,1200,200,500,450,3200,4200,1200,350]
sehwag=[1200,200,400,1100,200,30,3300,1400,200,1200]
kohli=[1100,100,100,2100,1500,250,3100,4000,3200,1350]
```

```python
x=np.arange(len(years))
width=0.25
plt.bar(x-width,sachin,width=width,label="Sachin Tendulkar")
plt.bar(x,sehwag,width=width,label="Sehwag")
plt.bar(x+width,kohli,width=width,label="Virat Kohli")
plt.xlabel("years")
plt.ylabel("Runs")
plt.title("Run comparison")
plt.xticks(x,years)
plt.legend()
```

```python
players=["Sachin","sehwag","Kohli","Yuvraj"]
runs_5yrs=[500+700+300+200+400,300+400+300+200+400,200+700+100+200+400,400+700+400+200+400]
```

```python
plt.barh(players,runs_5yrs,color="skyblue")
plt.xlabel("Total run in first 5 year")
plt.ylabel("First five year performance of indian batsman")
plt.tight_layout()
```
