```python
import matplotlib.pyplot as plt
```

```python
activities=['study','play','watch','sleep']
time_spent=[3,4,2,1]
colors=['blue','red','yellow','pink']
plt.figure(figsize=(6,6))
plt.pie(time_spent ,labels=activities ,colors=colors ,startangle=90,autopct="%1.1f%%")
plt.title("Activities of a day")
```

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
studying=[3,4,2,6,4,8,1]
playing=[2,2,4,1,6,3,1]
watching=[1,2,7,2,3,4,4]
sleeping=[5,6,4,3,7,3,8]
```

```python
labels=['Studying','Playing','TV','Sleeping']
colors=['blue','red','yellow','green']
plt.figure(figsize=(7,7))
plt.stackplot(days,studying,playing,watching,sleeping,labels=labels,colors=colors,alpha=0.9)
plt.legend(loc=("upper left"))
plt.title("Student Time Distribution")
plt.ylabel("Hours")
plt.xlabel("Days")
```
