```python
import matplotlib.pyplot as plt
```

```python
study_hours=[1,2,3,4,9,7,6,5,8]
exam_scores=[40,45,50,47,85,98,45,28,78]
```

```python
plt.scatter(study_hours,exam_scores)
```

```python
plt.scatter(study_hours,exam_scores)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Marks scored by students")
plt.title("Study vs Marks")
```

```python
colors=['red' if score<50 else 'green' for score in exam_scores]
sizes=[score*2 for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=colors,s=sizes)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Marks scored by students")
plt.title("Study vs Marks")
```

```python
#colors=['red' if score<50 else 'green' for score in exam_scores]
sizes=[score*2 for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='viridis')
plt.colorbar(label='score')
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Marks scored by students")
plt.title("Study vs Marks")
```

```python
#colors=['red' if score<50 else 'green' for score in exam_scores]
sizes=[score*2 for score in exam_scores]
social_score=[12,34,67,55,678,145,56,999,445]
plt.scatter(study_hours,exam_scores,c=social_score,cmap='viridis')
plt.colorbar(label='score')
plt.grid(True)
for i in range(len(study_hours)):
    plt.annotate(f's{i+1}',(study_hours[i]+0.2,exam_scores[i]))
plt.xlabel("Study Hours")
plt.ylabel("Marks scored by students")
plt.title("Study vs Marks")
```
